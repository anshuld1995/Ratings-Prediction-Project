# Scrapping Ratings and Reviews for Headphones from Amazon

```python
driver = webdriver.Chrome(r"C:\Users\anshul dubey\Documents\driver\chromedriver.exe")
```

```python
driver.maximize_window()
```

```python
driver.get("https://www.amazon.in/")
```

```python
search_bar = driver.find_element_by_xpath("//input[@id='twotabsearchtextbox']")
search_bar.send_keys("headphones")
```

```python
search_click = driver.find_element_by_xpath("//input[@id='nav-search-submit-button']")
search_click.click()
```

```python
# Creating empty lists
Product_URL=[]
Ratings=[]
Review=[]

#Getting URLs of the product
for i in range(1,4):
    URL = driver.find_elements_by_xpath('//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in URL:
        Product_URL.append(i.get_attribute('href'))
      
    try:
        next_btn=driver.find_element_by_xpath('//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]').click()
    except NoSuchElementException:
        pass
    
for i in Product_URL:
    driver.get(i)
    
    # Clicking the rating
    try:
        driver.find_element_by_xpath("//a[1][@id='acrCustomerReviewLink']").click()
    except NoSuchElementException:
        print("No rating")
        pass
    
    #Clicking to see all reviews
    try:
        driver.find_element_by_xpath("//a[@class='a-link-emphasis a-text-bold']").click()
    except NoSuchElementException:
        pass
    
     #Scrapping the details
    Start_page=1
    End_page=50
    for page in range(Start_page,End_page+1):
        try:
            Reviews=driver.find_elements_by_xpath("//div[@class='a-row a-spacing-small review-data']/span/span")
            for r in Reviews:
                Review.append(r.text.replace('\n',''))
        except NoSuchElementException:
            Review.append("Not Available")
        try:
            Rating=driver.find_elements_by_xpath("//div[@class='a-section celwidget']/div[2]/a[1]")
            for i in Rating:
                rating=i.get_attribute('title')
                Ratings.append(rating[:3])
        except NoSuchElementException:
            Ratings.append("Not available")      
        
        #Looping for going to next page automatically
        try:
            next_page=driver.find_element_by_xpath("//div[@id='cm_cr-pagination_bar']/ul/li[2]/a")
            if next_page.text=='Next Page':
                next_page.click()
                time.sleep(2)
        except NoSuchElementException:
            pass
```

```python
print(len(Review),len(Ratings))
```

```python
headphones=pd.DataFrame({'Product_Review':Review[:11000],'Ratings':Ratings[:11000]})
headphones
```

# Scrapping ratings and reviews for Router from FlipKart
